```python
#So we filter out all non-numeric year values
filtered_df = df_movies_with_director[df_movies_with_director['year'].str.isnumeric()]
```

```python
#And only keep English language movies after 1990
english_movies_after_1990 = filtered_df[(pd.to_numeric(filtered_df['year']) >= 1990) & (filtered_df['country'] == "USA")]
```
