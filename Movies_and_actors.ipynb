```python
movies_with_good_ranking_graph = ActorGraph()
good_ranking_movies_actors_list = [] # the list our movies and each list is an actor list for that movie
```

```python
df_imdb_movies["orig_lang"][1]
```

```python
from tqdm import tqdm
from actor_graph import remove_characters_from_crew_list
for i in tqdm(range(len(df_imdb_movies["crew"]))):
    if isinstance(df_imdb_movies["crew"][i], str) and df_imdb_movies["score"][i] >= 75.0 and df_imdb_movies["orig_lang"][i] == ' English':
        good_ranking_movies_actors_list.append(remove_characters_from_crew_list(df_imdb_movies["crew"][i]))
```

```python
good_ranking_movies_actors_list
```

```python
good_ranking_movies_actors_list_only_leaver_first_4 = []
for movie in tqdm(good_ranking_movies_actors_list):
    good_ranking_movies_actors_list_only_leaver_first_4.append(movie[:4])
```

```python
good_ranking_movies_actors_list_only_leaver_first_4
```

```python
movies_with_good_ranking_graph.make_graph(good_ranking_movies_actors_list_only_leaver_first_4)
```

```python
from matplotlib import pyplot as plt
import networkx as nx
pos = nx.forceatlas2_layout(movies_with_good_ranking_graph.graph, gravity=10)
plt.figure( 1, figsize=(10, 10))
nx.draw_networkx(movies_with_good_ranking_graph.graph, pos=pos, with_labels=False, node_size=10, alpha=0.3)
```
