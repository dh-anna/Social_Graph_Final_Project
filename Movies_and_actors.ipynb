```python
from importing_dataset import import_dataset, saving_df_to_pickle, loading_df_from_pickle

df_imdb_movies = loading_df_from_pickle('imdb-movies-dataset')
```

```python
from tqdm import tqdm
from actor_graph import remove_characters_from_crew_list
movies_actors_list = [] # the list our movies and each list is an actor list for that movie
for i in tqdm(range(len(df_imdb_movies["crew"]))):
    if isinstance(df_imdb_movies["crew"][i], str):
        movies_actors_list.append(remove_characters_from_crew_list(df_imdb_movies["crew"][i]))
```

```python
actors_graph = ActorGraph()
actors_graph.make_graph(movies_actors_list)
```

```python
type(actors_graph.graph.nodes())
```

```python
from matplotlib import pyplot as plt
import networkx as nx
pos = nx.forceatlas2_layout(actors_graph.graph, gravity=10)
plt.figure( 1, figsize=(10, 10))
nx.draw_networkx(actors_graph.graph, pos=pos, with_labels=False, node_size=10, alpha=0.3)
```
